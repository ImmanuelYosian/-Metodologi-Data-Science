3 Menggunakan Barchart karena dalam bar chart kita bisa melihat data dari jumlah yang tertinggi hingga terendah sehingga bisa digunakan dalam mengurutkan populasi relatif makanan penutup
makanan penutup yang harus disingkirkan adalah Puding Vanila,Pastel,Kue wajik karena ketiga makanan tersebut memiliki jumlah penjualan terendah dari urutan 1,2,3

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

nama_makanan = ('Donat', 'Pastel', 'Kue Coklat', 'Ice Cream', 'Puding Vanila', 'Brownies', 'Puding Strawberi', 'Puding Coklat', 'Ice Cream Nutela', 'Kue Coklat-Keju', 'Kue Wajik', 'Kue Sus', 'Mochi')
jumlah_penjualan = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({'Nama' :nama_makanan, 'Jumlah' : jumlah_penjualan,})
df.sort_values(by='Jumlah', inplace = True, ascending=False)

kor_x = np.arange(len(df))
warna = ['#6A5ACD' for _ in range(len(df))]
warna [-3] = '#FF0000'
warna [-2] = '#FF0000'
warna [-1] = '#FF0000'

plt.figure(figsize=(20,10))

plt.bar(kor_x, df['Jumlah'], tick_label=df['Nama'], color=warna)
plt.xticks(rotation=90)
plt.ylabel('Jumlah')
plt.title('Penjualan makanan')
plt.show() 
```
